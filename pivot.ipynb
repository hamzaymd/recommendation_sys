```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('DATA\clean_books2.xlsx')
rf = pd.read_excel('DATA\clean_rating2.xlsx')
```

```python
rf.rename(columns={ 'Book-Rating': 'rating'}, inplace=True)
```

```python
book_pivot = rf.pivot_table(columns='User-ID', index='Title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

TRAINING MODEL

```python
from scipy.sparse import csr_matrix
```

```python

book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python

model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[208,:].values.reshape(1,-1), n_neighbors=6 )
distance
```

TESTING MODEL

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = '"Harry Potter and the Chamber of Secrets (Book 2)"'
recommend_book(book_name)
```
